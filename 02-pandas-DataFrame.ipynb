```python
import pandas as pd
import numpy as np

data = {
    'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
    'year': [2000, 2001, 2002, 2001, 2002, 2003],
    'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]
}
frame = pd.DataFrame(data)

## 取前5行
frame.head()
```

```python
# 指定列序列
pd.DataFrame(data, columns=['year', 'state', 'pop'])
```

```python
frame2 = pd.DataFrame(data, columns=['year', 'state', 'pop', 'debt'],
   ....:                       index=['one', 'two', 'three', 'four',
   ....:                              'five', 'six'])

frame2

# 取出为 Series
frame2.year
```

```python
# index 取数
frame2.loc['three']
frame2.loc['one']
```

```python
# 列赋值
frame2.debt = 3.1415926
frame2.debt = np.arange(6.)
frame2


# 索引 Series 赋值
val = pd.Series({'two': 111, 'five': 555, 'three': 333 })

frame2.debt = val
frame2
```

```python
# 为不存在的列赋赋值会创建新的一列
frame2['foo'] = frame2.state == 'Ohio'
frame2

# del 可以的删除一列
del frame2['foo']
frame2
```
